```python
#define features

def read_data(filename):
    X = []
    Y = []
    with open(filename, encoding = 'utf-8') as f:
        for l in f:
            cols = l.split('\t')
            X.append(cols[1])
            Y.append(cols[0])
    return X, Y
Xtrain, Ytrain = read_data('dataset/a2_first_sample.tsv')
print(Xtrain)
```

```python
import re
tokenize_re = re.compile(r'''
                         \d+[:\.]\d+
                         |(https?://)?(\w+\.)(\w{2,})+([\w/]+)
                         |[@\#]?\w+(?:[-']\w+)*
                         |[^a-zA-Z0-9 ]+''', re.VERBOSE)

def tokenize(text):
    return [ m.group() for m in tokenize_re.finditer(text) ]
```

```python
Xtrain_tokenized = []

for comment in Xtrain:
    Xtrain_tokenized.append(tokenize(comment))
#print(Xtrain_tokenized)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

Xtrain_tok_Vect = []

vectorizer = CountVectorizer()
Xtrain_tok_Vect = vectorizer.fit_transform(Xtrain)

print(vectorizer.get_feature_names())

analyze = vectorizer.build_analyzer()
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction import DictVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import Perceptron
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.dummy import DummyClassifier

pipeline = make_pipeline(
    TfidfVectorizer(),
    Perceptron()
)
```

```python
from sklearn.model_selection import cross_validate
import warnings
warnings.filterwarnings('ignore')

results  = cross_validate(pipeline, Xtrain, Ytrain)
print(results)
```
